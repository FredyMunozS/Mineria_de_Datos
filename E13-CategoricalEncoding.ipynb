```python
num_df = df.select_dtypes(include=['float64','int64']).copy()
num_df.head()
```

Para completar la variable normalized_losses, se infiere a travÃ©s del valor correspondiente a vechiculos de la misma marca y engine_size que si tengan este dato. Con esta tÃ©cnica se pueden completar 32 observaciones, las 10 restantes se hacen por la mediana.

```python
df2=df.groupby(['make', 'normalized_losses', 'engine_size'])['engine_size'].count().to_frame(name='Cantidad').reset_index()
df2=df2.loc[df2.engine_size.duplicated(keep='first'), :]
```

```python
df2.head()
```

```python
df['normalized_losses']=df['normalized_losses'].mask(pd.isnull, pd.merge(df, df2, on=['engine_size','make' ], how='left')['normalized_losses_y'])
df.normalized_losses.fillna(df.normalized_losses.median(), inplace=True)
```

Se aplica la misma tÃ©cnica para las siguientes variables: bore, stroke, horsepower, peak_rpm, price

```python
df2=df.groupby(['make', 'bore', 'engine_size'])['engine_size'].count().to_frame(name='Cantidad').reset_index()
df2=df2.loc[df2.engine_size.duplicated(keep='first'), :]

df['bore']=df['bore'].mask(pd.isnull, pd.merge(df, df2, on=['engine_size','make' ], how='left')['bore_y'])
df.bore.fillna(df.bore.median(), inplace=True)
```

```python
df2=df.groupby(['make', 'stroke', 'engine_size'])['engine_size'].count().to_frame(name='Cantidad').reset_index()
df2=df2.loc[df2.engine_size.duplicated(keep='first'), :]

df['stroke']=df['stroke'].mask(pd.isnull, pd.merge(df, df2, on=['engine_size','make' ], how='left')['stroke_y'])
df.stroke.fillna(df.stroke.median(), inplace=True)
```

```python
df2=df.groupby(['make', 'horsepower', 'engine_size'])['engine_size'].count().to_frame(name='Cantidad').reset_index()
df2=df2.loc[df2.engine_size.duplicated(keep='first'), :]

df['horsepower']=df['horsepower'].mask(pd.isnull, pd.merge(df, df2, on=['engine_size','make' ], how='left')['horsepower_y'])
df.horsepower.fillna(df.horsepower.median(), inplace=True)
```

```python
df2=df.groupby(['make', 'peak_rpm', 'engine_size'])['engine_size'].count().to_frame(name='Cantidad').reset_index()
df2=df2.loc[df2.engine_size.duplicated(keep='first'), :]

df['peak_rpm']=df['peak_rpm'].mask(pd.isnull, pd.merge(df, df2, on=['engine_size','make' ], how='left')['peak_rpm_y'])
df.peak_rpm.fillna(df.peak_rpm.median(), inplace=True)
```

```python
df2=df.groupby(['make', 'price', 'engine_size'])['engine_size'].count().to_frame(name='Cantidad').reset_index()
df2=df2.loc[df2.engine_size.duplicated(keep='first'), :]

df['price']=df['price'].mask(pd.isnull, pd.merge(df, df2, on=['engine_size','make' ], how='left')['price_y'])
df.price.fillna(df.price.median(), inplace=True)
```

# Exercise 13.2

Split the data into training and testing sets

Train a Random Forest Regressor to predict the price of a car using the nominal features

```python
y = df['price']
categorical=['price','make','fuel_type','aspiration','num_doors','body_style','drive_wheels'
             ,'engine_location','engine_type','num_cylinders','fuel_system']
X = df.drop(categorical, axis=1)
```

```python
# train/test split
from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(X_.iloc[samples, :], y.iloc[samples], random_state=1, test_size=0.5)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.3)
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
clf = RandomForestRegressor(n_jobs=-1)
clf.fit(X_train, y_train)
y_pred_df = clf.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_pred_df, y_test))
```
